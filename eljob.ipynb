```python
import re
import requests
from bs4 import BeautifulSoup

OUTPUT = 'console'

url = 'https://eljob.ru/test/137_'

def string_handler(string):
    res_string = re.sub('(^[\t\s]+|[\t\s\n]+$)', '', string)
    res_string = re.sub('\s{2,}', ' ', res_string)
    
    return res_string

def task_handler(task):    
    return\
        ('  ', '* ')['correct' in task['class']] +\
        string_handler(task.text)

global_string = '';

for i in range(1, 19):
    page = requests.get(url + str(i))
    html = BeautifulSoup(page.text)
    
    test_string =\
        '\n' +\
        f'[ TEST {i} ]' + '\n'
    
    for task_container in html.findAll('div', 'item'):
        test_string +=\
            '\n' +\
            string_handler(task_container.find('pre', 'title').text) + '\n' +\
            '\n' +\
            '\n'.join(
                map(task_handler, task_container.findAll('div', 'title'))) +\
            '\n'
        
    if OUTPUT == 'console':
        print(test_string)
     
    if OUTPUT == 'file':
        with open(f'tests/{i}.txt', 'w') as file:
            file.write(test_string)
            file.close()
    
    global_string += test_string
    
if OUTPUT == 'file':
    with open('tests/all-in-one.txt', 'w') as file:
        file.write(global_string)
        file.close()
```
